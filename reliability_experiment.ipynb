```python
import sklearn as sk
import pandas as pd
import numpy as np
```

```python
from itertools import product, combinations_with_replacement
```

```python
def parallel_r(plist):
    return 1 - np.prod([1 - p for p in plist])

def system_r_simple(i):
    netR = parallel_r(i[2])
    nodes = [i[1] * netR,] * i[0]
    return parallel_r(nodes)

def system_r_simple_vec(nodes, nodeR, net_plist):
    netR = parallel_r(net_plist)
    return parallel_r([nodeR * netR] * nodes)

def complex(i):
    nets = sum(bool(n) for n in i[2])
    return (1 + nets) + (i[0] * nets)
```

```python
N = np.random.randint(1,20)
M = np.random.randint(1,6)
k = np.random.randint(1,4)

idx = pd.Index(list(product(list(range(1,N)),
                           [x/10 for x in range(1,10)],
                           list(combinations_with_replacement([x/10 for x in range(10)],M))
                          )),
              names=['nodes','nodeR','networks'])
```

```python
#data = idx.to_frame()
data = pd.DataFrame(index=idx)
data2 = data.index.to_frame()
```

```python
%%time
data['systemR'] = data.index.map(system_r_simple)
```

```python
%%time
data2['systemR_vec'] = system_r_simple_vec(data2.nodes, data2.nodeR, data2.networks)
```

```python
data['complex'] = data.index.map(complex)
data['complex_norm'] = (data.complex - data.complex.min()) / (data.complex.max() - data.complex.min())
```
